## Linear Regression Basic Intution

Regression analysis is a form of predective modelling technique which investigates the relationship between a dependent and independent variable. It includes graphing a line over set of data points that best fits the data. Here Y is a dependent variable whereas x is a independent variable.


Three major uses of Linear regression:
1. Determining the strength of the predictors
2. Forcasting an effect
3. Trend forecasting

#### Basic difference between Linear Regression and logistic regression:

Where the Linear regression is used:
1. Evaluating trends and sales estimate
2. Analyzing the impact of price changes
3. Assesment of risks in financial services and insurance domain

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
data = pd.read_csv('headbrain.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
x = data['Head Size(cm^3)'].values
y = data['Brain Weight(grams)'].values
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)
```

```python
mean_x
```

```python
mean_y
```

```python
n = len(x)

num = 0 
den = 0
for i in range(n):
    num += (x[i] - mean_x) * (y[i] - mean_y)
    den += (x[i] - mean_x) ** 2
    
m = num / den
c = mean_y - (m * mean_x)

print(m, c)
```

In the equation of line y=mx+c
<br>
y = brain weight<br>
m = 0.263<br>
x = head size<br>
c = 325.573

```python
max_x = np.max(x) + 100
min_x = np.min(x) - 100

x_ = np.linspace(min_x, max_x, 1000)
y_ = c + m*x_

plt.plot(x_, y_, label='Regression Line')

plt.scatter(x, y, label='Scatter Plot')

plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')

plt.legend()
plt.show()
```

```python
ss_t = 0
ss_r = 0
for i in range(n):
    y_pred = c + m * x[i]
    ss_t += (y[i] - mean_y) ** 2
    ss_r += (y[i] - y_pred) ** 2
    
r2_val = 1-(ss_r/ss_t)
print(r2_val*100)
```
